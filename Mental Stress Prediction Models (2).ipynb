## Checking Feature Importance By using Statistical Test

#### Applying the Chi-Squared test & P-test  on the category features for checking the features importance to target
Correlation with Target Variable:

Analyze the relationship between each categorical feature and the target variable ('PerformanceRating'). You can use tools like Chi-square test or Cramér's V for association.

```python
from scipy.stats import chi2_contingency

# List to store results
chi2_results = []

for column in data.columns[:-1]:  # Exclude the target variable
    contingency_table = pd.crosstab(data[column], data['Target'])
    
    # Perform Chi-squared test
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    
    # Store results
    chi2_results.append({'Feature': column, 'Chi2': chi2, 'P-value': p_value})

# Display the results
chi2_results_df = pd.DataFrame(chi2_results)
print(chi2_results_df)
```

Insights: The Chi-squared (Chi2) statistic measures the independence between two categorical variables. In the context of your analysis, you are comparing each categorical feature (e.g., Gender, Age, Blank_mind, etc.) with the target variable Target. The Chi2 statistic and its associated p-value help you determine whether there is a statistically significant association between the categorical feature and the target variable.

Here's how to interpret the Chi2 statistic:

1) Higher Chi2 Value: A higher Chi2 value indicates a stronger association between the categorical feature and the target variable.

2) Lower Chi2 Value: A lower Chi2 value suggests a weaker association.

In your results:

For example, Anxiety_Personal vs Target - Chi2: 125.08 indicates a relatively high Chi2 value, suggesting a strong association between features and the target variable.

On the other hand, Gender vs PerformanceRating - Chi2: 0.23 has a very low Chi2 value, indicating a weaker association.

To assess the statistical significance of these associations, you would typically compare the Chi2 value to a critical value from the Chi-squared distribution or, more commonly, look at the associated p-value. The p-value helps determine whether the observed association is likely due to chance.

1) Significant p-value (typically < 0.05): Indicates that the association is statistically significant, and you may reject the null hypothesis of independence.

2) Non-significant p-value (typically >= 0.05): Suggests that the association is not statistically significant, and you may fail to reject the null hypothesis.

Let's look at a few examples from your results:

Anxiety_Personal: P-value is very close to zero (2.947338e-23), indicating a highly significant association with the target variable.

Gender: P-value is 0.0.2, which is higher than 0.05. This suggests that there is no significant association between Gender and the target variable.

## Logistic Regression

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
lr= OneVsRestClassifier(LogisticRegression())
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)

y_preda = lr.predict_proba(X_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,f1_score,precision_score,recall_score,roc_auc_score

print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
lr_acc_score= accuracy_score(y_test,y_pred)
lr_precision = precision_score(y_test,y_pred, average='weighted')
lr_recall = recall_score(y_test,y_pred, average='weighted')
lr_f1_score = f1_score(y_test,y_pred, average='weighted')

lr_cv_score = cross_val_score(lr, X_scaled, y, cv=3, scoring = 'accuracy').mean()
lr_AUC = roc_auc_score(y_test, y_preda, multi_class='ovr')
```

The Logistic Regression model gives 44% accuracy

```python
from sklearn.metrics import auc,roc_curve
from sklearn.preprocessing import label_binarize

#binarize y_values
classes = np.unique(y_test)
y_bin = label_binarize(y_test,classes=classes)

#roc curve for classes
fpr = {}
tpr = {}
threshold = {}
roc_auc = {}

n_class = classes.shape[0]

fig=plt.figure(figsize=(8,6))
for i in range(n_class):
    fpr[i],tpr[i],threshold[i] = roc_curve(y_bin[:,i], y_preda[:,i])
    roc_auc[i] = auc(fpr[i], tpr[i])

    #plotting
    plt.plot(fpr[i], tpr[i], linestyle='--',
            label='%s Vs Rest (AUC=%0.2f)'%(classes[i],roc_auc[i]))


plt.plot([0,1],[0,1],'b--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.title('Multiclass Roc Curve for LG')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```

## ROC Curves For Models


 ROC Curves for the Receiver Operating Characteristic (ROC) Curve is a model selection probability curve and plots the true positive rate against the false positive rate. ROC plot provides:

       Relationship between sensitivity and specificity
       1)accuracy: closer graph is to top and left-hand borders, the more accurate the test.
       2)accuracy: closer graph to diagonal, the less accurate the test.
       3)perfect test would go straight from zero up to top-left corner and then straight across the horizontal
       4)prefect chance would follow the diagonal line and represent random outcome, i.e. flip of a coin
       5)useless test is any ROC plot under the diagonal line

```python
from sklearn.metrics import auc,roc_curve
from sklearn.preprocessing import label_binarize


#  ROC plot
c = 1  # subplot counter
fig = plt.figure(figsize=(50,50))

for name, model in models:
    model.probability = True   # need for predict_proba to work
    model.fit(X_train,y_train)
    y_preda = model.predict_proba(X_test)
    y_pred = y_preda[:,1]   # positive values only

    #binarize y_values
    classes = np.unique(y_test)
    y_bin = label_binarize(y_test,classes=classes)

    #roc curve for classes
    fpr = {}
    tpr = {}
    threshold = {}
    roc_auc = {}

    n_class = classes.shape[0]

    fig=plt.figure(figsize=(10,6))

    for i in range(n_class):
      fpr[i],tpr[i],threshold[i] = roc_curve(y_bin[:,i], y_preda[:,i])
      roc_auc[i] = auc(fpr[i], tpr[i])

      #plotting
      plt.plot(fpr[i], tpr[i], linestyle='--',
               label='%s Vs Rest (AUC=%0.2f)'%(classes[i],roc_auc[i]))
      plt.plot([0,1],[0,1],'b--')
      plt.xlim([0,1])
      plt.ylim([0,1])
      plt.title("Multiclass ROC Curve For {}".format(name), fontsize = 16)
      plt.xlabel('False Positive Rate')
      plt.ylabel('True Positive Rate')
      plt.legend(loc='lower right')

      c = c + 1

plt.tight_layout()
plt.show()
```
